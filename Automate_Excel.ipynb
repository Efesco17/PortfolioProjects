Create percentage change columns for each column.
To do that we need to reorder the data in ascending rather than descending order.

```python
data=data.sort_index()
```

```python
data.head()
```

```python
data['rev%-change']=data['revenue'].pct_change()
```

```python
data['eps%-change']=data['eps'].pct_change()
```

```python
data['fcf%-change']=data['fcf'].pct_change()
```

Create a new dataset containing only data from the last 5 years

```python
data_set=data.loc['2016-01-01':]
data_set.index=data_set.index.year
data_set
```

Create different dataframes for revenue,eps and fcf

```python
rev=data_set[['revenue','rev%-change']].copy()
eps=data_set[['eps','eps%-change']].copy()
fcf=data_set[['fcf','fcf%-change']].copy()
```

```python
def create_sheet(writer,df,sheet_name):
    df.to_excel(writer, sheet_name=sheet_name)
    worksheet=writer.sheets[sheet_name]
    workbook=writer.book
    
    # Create a line chart
    chart = workbook.add_chart({'type': 'line'})
    chart.add_series({
        'name':sheet_name,
        'categories':[sheet_name,1,0,5,0],
        'values':[sheet_name,1,1,5,1]
    })
    
    # Create a column chart for percentage change column
    column_chart=workbook.add_chart({'type':'column'})
    # Configure the series of the chart from the dataframe data.
    column_chart.add_series({
        'name': sheet_name,
        'categories':[sheet_name,1,0,5,0],
        'values':[sheet_name,1,2,5,2],
        'gap':100,
        'y2_axis': True,
    })
    
    column_chart.set_y2_axis({'name':'%-chg'})
    # Combine the two charts to make one chart
    chart.combine(column_chart)
 
    # Insert the chart into an actual excel cell
    worksheet.insert_chart('E1',chart)
    
 
 
```

```python
writer=pd.ExcelWriter('financials.xlsx', engine='xlsxwriter')
workbook=writer.book

create_sheet(writer,rev,'Revenue')
create_sheet(writer,eps,'EPS')
create_sheet(writer,fcf,'FCF')

writer.save()
workbook.close
```
